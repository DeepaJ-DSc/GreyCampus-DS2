# Plots in Python

```python
import matplotlib.pyplot as plt
```

## Bar Chart

```python
x = ["Asia","Africa","Australia","Europe","North America","South America"]
y = [104, 89, 37, 52, 167, 68]

plt.bar(x,y)
plt.show()
```

## Pie Chart

```python
plt.pie(y,labels=x)
# plt.legend()
plt.show()
```

## Histogram

```python
import numpy as np
import random

x = np.random.normal(170,10,250)
print (x)
```

```python
plt.hist(x)
plt.show()
```

## Line plot

```python
a= [1,2,3,4,5,6]
b= [20,84,37,98,64,48]
plt.plot(a,b)
plt.show()
```

## Scatter plot

```python
plt.scatter(a,b)
plt.show()
```

## Box and Whisker Plot

```python
x = np.random.normal(100,20,250)
print (x)
```

```python
plt.boxplot(x)
plt.show()
```
