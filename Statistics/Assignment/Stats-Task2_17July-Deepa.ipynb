## Sampling Methods

#### Importing data

```python
import numpy as np
import pandas as pd

data= pd.read_csv("F:/GreyCampus-Datascience/DS5/Statistics/Fakeschool_1.csv")
data
```

### 1. Simple Random Sampling

```python
# Generate a random sample of size 7
sample_random= data.sample(n=7)
sample_random
```

### 2. Systematic Random Sampling

```python
# Systematic Sampling: Choose Samples at regular intervals

index = np.arange(0,len(data),step=4)
sys_sample = data.iloc[index]
sys_sample
```

### 3. Cluster Sampling

```python
# Create clusters based on Class and consider all the data points from random samples

import random

c1= data[data['class']=='Fr']
c2= data[data['class']=='Jr']
c3= data[data['class']=='Sr']
c4= data[data['class']=='So']

clus_sample = random.choice(['c1','c2','c3','c4'])

if (clus_sample == 'c1'):
    print(c1)
elif(clus_sample == 'c2'):
    print(c2)
elif(clus_sample == 'c3'):
    print(c3)
else:
    print(c4)
```

### 4. Stratified sampling

```python
# Create different strata based on 'class' and draw random samples from each strata

s1_sample = c1.sample(n=2)
s2_sample = c2.sample(n=2)
s3_sample = c3.sample(n=2)
s4_sample = c4.sample(n=2)

strat_sample = s1_sample.append([s2_sample,s3_sample,s4_sample])
strat_sample
```
